```python
# 주어진 코드
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p1_1_.csv')
```

```python
# 1. 20L가격과 5L가격이 모두 0원이 아닌 데이터만 필터를 한 후, 각 row별로 20L가격과 5L가격의 차이를 ‘차이가격’ 이라 부른다고 하자. 시도명 별 차이가격의 평균가격을 비교할때 그 값이 가장 큰 금액을 반올림하여 소숫점 이하 1자리까지 구하여라
df1 = df[(df['20L가격'] != 0) & (df['5L가격'] != 0)].reset_index(drop=True)

df1['차이가격'] = df1['20L가격'] - df1['5L가격']


f = df1.groupby(['시도명'])['차이가격'].mean().sort_values().to_frame()
print(round(f['차이가격'].max() , 1))
```

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p1_2_.csv')
df.head(5)

#2. BMI는 몸무게(kg) / (키(M) * 키(M)) 로 정의 된다. 초고도 비만은 BMI 25이상 , 고도 비반은 BMI 25미만 - 23이상 , 정상은 23미만 - 18.5이상 저체중은 18.5미만으로 정의 된다. 주어진 데이터에서 초고도비만 인원 + 저체중 인원 의 숫자는?
df['height(m)'] = df['height(cm)'] * 0.01
df['BMI'] = df['weight(kg)'] / (df['height(m)'] * df['height(m)'])

over = df[df['BMI'] >= 25].shape[0]
under = df[df['BMI'] < 18.5 ].shape[0]

print(over + under)
```

#### 작업 2유형
벤츠 차량 가격 예측 : 예측 변수 price, test.csv에 대해 ID별로 price 값을 예측하여 제출, 제출 데이터 컬럼은 ID와 price 두개만 존재해야함. 평가지표는 rmse


```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p2_train_.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p2_test_.csv')

test.head()
train.head()

train.isnull().sum()
test.isnull().sum()
#결측치 없음
```

```python
# 데이터 전처리

y = train['price']
x_train = train.drop(columns =['price','ID'])
x_test = test.drop(columns =['ID'])

x_train_dum = pd.get_dummies(x_train)
x_test_dum = pd.get_dummies(x_test)

#더미화 해주고나면 train셋과 test셋 컬럼 개수가 차이남 , reindex 해줘야함 !!!
x_test_dum = x_test_dum.reindex(columns = x_train_dum.columns, fill_value=0)
```

```python
# import 및 모델 생성
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor
import numpy as np

X_train,X_test,Y_train,Y_test = train_test_split(x_train_dum,y)


rr = RandomForestRegressor()
rr.fit(X_train,Y_train)


pred = rr.predict(X_test)
print('test rmse', np.sqrt(mean_squared_error(Y_test,pred)))


pred_test = rr.predict(x_test_dum)

sumission = pd.DataFrame()
sumission['ID'] = test['ID']
sumission['price']  =pred_test

sumission.head()
# sumission.to_csv('00000000.csv',index=Fasle)
```

#### 작업 3유형



```python
import pandas as pd
df= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p3_1.csv')
df.head()
```

```python
#1. 55명 학생들의 키에 대한 표본 평균을 구하여라(반올림하여 소숫점 3째자리까지
import numpy as np
mean = round(np.mean(df.height),3)
print(round(np.mean(df.height),3))
```

```python
#2. t분포 양쪽 꼬리에서의 t 값을 구하여라 (반올림하여 소수4째자리까지)
from scipy.stats import t

std = np.std(df.height, ddof = 1)
n = 55

# 신뢰수준, 자유도
con = 0.95
ddof = n-1

# t분포
t_value = round(t.ppf((1+con) / 2 , ddof),4)
print(t_value)
```

```python
#3. 95% 신뢰구간을 구하여라

lower = round(mean-t_value*std / np.sqrt(n),3)
upper = round(mean+t_value*std / np.sqrt(n),3)
print(lower,upper)
```
