주어진 데이터(basic2.csv)에서 주 단위 Sales의 합계를 구하고, 가장 큰 값을 가진 주와 작은 값을 가진 주의 차이를 구하시오(절대값)

```python
import pandas as pd

df = pd.read_csv('/content/basic2.csv', parse_dates=['Date'], index_col=0)
```

```python
print(df.shape)
df.head()
```

```python
# 주 단위로 나누기
df_w = df.resample('W').sum()
df_w.head()
```

```python
# 결과 출력
print(df_w['Sales'].max() - df_w['Sales'].min())
```
