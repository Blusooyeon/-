### 이상치를 찾아라 (IQR 활용)
titanic 데이터에서 IQR 을 이용해 Fare 컬럼의 이상치를 찾고, 이상치 데이터의 여성 수를 구하시오.


```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
import numpy as np

df = pd.read_csv('/content/train.csv')
```

```python
# 간단한 EDA
print(df.shape)
print(df.isnull().sum())
df.head()
```

```python
# IQR 구하기
Q1 = df['Fare'].quantile(.25)
Q3 = df['Fare'].quantile(.75)

IQR = Q3 - Q1
Q1 - 1.5*IQR , Q3 + 1.5*IQR
```

```python
# 이상치 데이터 구하기
outdata1 = df[df['Fare'] < -26.724]
outdata2 = df[df['Fare'] > 65.6344]

len(outdata1), len(outdata2)
```

```python
# 이상치 데이터 중에 여성 수 구하기
print(sum(outdata2['Sex'] == 'female'))
```

✅Check   
1. quantile(.25) , quantile(.75) -->  1분위수, 3분위수 기억하기
2. 이상치 측정 범위  : Q1 - 1.5IQR ~ Q3 + 1.5IQR
