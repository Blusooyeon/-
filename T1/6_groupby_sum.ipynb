### 결측치 제거 및 그룹 합계
주어진 데이터 중 'f1' 컬럼 결측 데이터를 제거하고, 'city'와 'f2'을 기준으로 묶어 합계를 구하고, 'city가 경기이면서 f2가 0'인 조건에 만족하는 f1 값을 구하시오


```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
import numpy as np

df = pd.read_csv('/content/basic1.csv')
df.head()
```

```python
#f1 컬럼 결측치 제거
df = df[~df['f1'].isnull()]
```

```python
# 그룹 합계 계산
df2 = df.groupby(['city','f2']).sum()
df2
```

```python
# 조건에 맞는 값 출력
print(df2.iloc[0]['f1'])
```
