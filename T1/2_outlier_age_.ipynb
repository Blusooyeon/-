### 이상치를 찾아라 ( 소수점 나이 )     
주어진 데이터에서 이상치(소수점 나이)를 찾고, 올림,내림,버림 했을 때 3가지 모두 이상치 'age' 평균을 구한 다음 모두 더하여 출력하시

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
import numpy as np

df = pd.read_csv('/content/basic1.csv')
df.head()
```

```python
# 소수점 데이터 찾기
df = df[(df['age']-np.floor(df['age'])) != 0 ]
df
```

```python
# 이상치를 포함한 데이터 올림, 내림, 버림의 평균값
# 올림
m_ceil = np.ceil(df['age']).mean()
# 내림
m_floor = np.floor(df['age']).mean()
# 버
m_trunc = np.trunc(df['age']).mean()

m_ceil, m_floor, m_trunc
```

```python
# 평균값 더한 다음 출력
print(m_ceil + m_floor + m_trunc)
```

✅Check
1. 올림: np.ceil()      
내림: np.floor()      
버림: np.trunc()
