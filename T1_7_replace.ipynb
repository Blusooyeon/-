###  값 변경 및 2개 이상의 조건
'f4'컬럼 값이 'ESFJ'인 데이터를 'ISFJ'로 대체하고,      
'city'가 '경기'이면서 'f4'가 'ISFJ'인 데이터 중 'age' 컬럼의 최댓값을 출력하시오



```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/basic1.csv')
```

```python
# ESFJ 값을 가진 데이터 확인
df[df['f4'] == 'ESFJ']
```

```python
# 값 변경하기
df['f4'] = df['f4'].replace('ESFJ','ISFJ')
```

```python
# 두가지 조건을 만족하는 값
df2 = df[(df['f4']=='ISFJ') & (df['city']=='경기')]
```

```python
# age 컬럼의 최댓값
df2['age'].max()
```
