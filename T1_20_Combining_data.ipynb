### 고객과 잘 맞는 타입 추천
basic1 데이터와 basic3 데이터를 'f4'값을 기준으로 병합하고,        
병합한 데이터에서 r2결측치를 제거한다음, 앞에서 부터 20개 데이터를 선택하고 'f2'컬럼 합을 구하시오

```python
import pandas as pd

b1= pd.read_csv('/content/basic1.csv')
b3= pd.read_csv('/content/basic3.csv')
```

```python
b1.head()
```

```python
b3.head()
```

```python
# 데이터 결합
df = pd.merge(left=b1, right=b3, how='left', on='f4')
df.head()
```

```python
# 결측치 확인
df.isnull().sum()
```

```python
# r2 컬럼 결측치 제거
print(df.shape)
df = df.dropna(subset=['r2'])
print(df.shape)
```

```python
# 결과 출력
df = df.reset_index()

print(df.iloc[:20]['f2'].sum())
```
