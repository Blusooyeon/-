city와 f4를 기준으로 f5의 평균값을 구한 다음, f5를 기준으로 상위 7개 값을 모두 더해 출력하시오 (소수점 둘째자리까지 출력)

```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/basic2.csv')
df.head()
```

```python
#city와 f4 별 f5 의 평균 값 출력
df = df.groupby(['city','f4'])[['f5']].mean()
print(df)
```

```python
# df 전환 후 상위 7개 출력
df = df.reset_index().sort_values('f5', ascending=False).head(7)

# 결과 출력
round(df['f5'].sum(),2)
```
