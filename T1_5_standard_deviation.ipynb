### 조건에 맞는 데이터 표준편차 구하기
주어진 데이터 중 'f4' 컬럼 값이 'ENFJ'와 'INFP' 인 'f1'의 표준편차 차이를 절대값으로 구하시오

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
import numpy as np

df = pd.read_csv('/content/basic1.csv')
df.head()
```

```python
# 조건에 맞는 데이터 찾기
df[df['f4']=='ENFJ']
```

```python
# 조건에 맞는 f1의 표준편자 구하기
enfj = df[df['f4']=='ENFJ']['f1'].std()

infp = df[df['f4']=='INFP']['f1'].std()
```

```python
# 두 표준편차 차이 절댓값 출력
print(np.abs(enfj-infp))
```
