작업 2유형                 
벤츠 차량 가격 예측 : 예측 변수 price, test.csv에 대해 ID별로 price 값을 예측하여 제출, 제출 데이터 컬럼은 ID와 price 두개만 존재해야함. 평가지표는 rmse


```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p2_train_.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p2_test_.csv')

test.head()
train.head()

train.isnull().sum()
test.isnull().sum()
#결측치 없음
```

```python
# 데이터 전처리

y = train['price']
x_train = train.drop(columns =['price','ID'])
x_test = test.drop(columns =['ID'])

x_train_dum = pd.get_dummies(x_train)
x_test_dum = pd.get_dummies(x_test)

#더미화 해주고나면 train셋과 test셋 컬럼 개수가 차이남 , reindex 해줘야함 !!!
x_test_dum = x_test_dum.reindex(columns = x_train_dum.columns, fill_value=0)
```

```python
# import 및 모델 생성
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor
import numpy as np

X_train,X_test,Y_train,Y_test = train_test_split(x_train_dum,y)


rr = RandomForestRegressor()
rr.fit(X_train,Y_train)


pred = rr.predict(X_test)
print('test rmse', np.sqrt(mean_squared_error(Y_test,pred)))


pred_test = rr.predict(x_test_dum)

sumission = pd.DataFrame()
sumission['ID'] = test['ID']
sumission['price']  =pred_test

sumission.head()
# sumission.to_csv('00000000.csv',index=Fasle)
```

```python
import pandas as pd
df= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e5_p3_1.csv')
df.head()
```

```python
#1. 55명 학생들의 키에 대한 표본 평균을 구하여라(반올림하여 소숫점 3째자리까지
import numpy as np
mean = round(np.mean(df.height),3)
print(round(np.mean(df.height),3))
```

```python
#2. t분포 양쪽 꼬리에서의 t 값을 구하여라 (반올림하여 소수4째자리까지)
from scipy.stats import t

std = np.std(df.height, ddof = 1)
n = 55

# 신뢰수준, 자유도
con = 0.95
ddof = n-1

# t분포
t_value = round(t.ppf((1+con) / 2 , ddof),4)
print(t_value)
```

```python
#3. 95% 신뢰구간을 구하여라

lower = round(mean-t_value*std / np.sqrt(n),3)
upper = round(mean+t_value*std / np.sqrt(n),3)
print(lower,upper)
```

```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e4_p2_train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e4_p2_test.csv')

display(train.head(2))
test.head(2)
```

```python
# 결측값 처리
MAX_EVER_MARRIED = train['Ever_Married'].value_counts().index[0]
MAX_GRADUATED = train['Graduated'].value_counts().index[0]
MAX_PROFESSION = train['Profession'].value_counts().index[0]
MEAD_Work_Experience = train['Work_Experience'].mean()
MAX_Family_Size = train['Family_Size'].value_counts().index[0]
MAX_Var_1 = train['Var_1'].value_counts().index[0]

fillna_dic = {
    'Ever_Married':MAX_EVER_MARRIED,
    'Graduated':MAX_GRADUATED,
    'Profession':MAX_PROFESSION,
    'Work_Experience':MEAD_Work_Experience,
    'Family_Size':MAX_Family_Size,
    'Var_1':MAX_Var_1
}

for col in fillna_dic.keys():
    train.loc[:,col] = train.loc[:,col].fillna(fillna_dic[col])
    test.loc[:,col] = test.loc[:,col].fillna(fillna_dic[col])

```

```python
# Family_Size 범주형 변경
train['Family_Size'] = train['Family_Size'].astype('str')
test['Family_Size'] = test['Family_Size'].astype('str')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
from sklearn.ensemble import RandomForestClassifier


train_drop = train.drop(columns = ['ID','Segmentation'])
test_drop = test.drop(columns = ['ID'])
y = train['Segmentation']


train_dum = pd.get_dummies(train_drop)
test_dum = pd.get_dummies(test_drop)[train_dum.columns]

x_train, x_test, y_train,y_test = train_test_split(train_dum , y ,random_state =1,stratify=y)

rf = RandomForestClassifier(random_state =1)
rf.fit(x_train,y_train)

pred = rf.predict(x_test)
print('valiation test f1score',f1_score(y_test,pred,average='macro'))


pred_test = rf.predict(test_dum)


sumission = pd.DataFrame()
sumission['ID'] = test['ID']
sumission['Segmentation'] = pred_test
sumission.to_csv('000000000.csv',index=False)
```

```python
#1.학과 평균 인원에 대한 값을 소숫점 이하 3자리까지 구하여라
round(df.groupby(['학과']).size().mean() ,3 )
```

```python
#2. 카이제곱검정 독립성 검정 통계량을 소숫점 이하 3자리까지 구하여라

from scipy.stats import chi2_contingency
df1 = pd.crosstab(df['학과'],df['성별'])
chi2_contingency(df1)
```

```python
# 어느 학교에서 수학 시험을 본 학생 100명 중 60명이 60점 이상을 받았다. 이 학교의 수학 시험의 평균 점수가 50점 이상인지 95%의 신뢰 수준에서 검정하려한다.
import numpy as np

n = 100
p_hat = 0.6
p = 0.5
alpha = 0.05

# 검정 통계량 계산
z = round((p_hat - p) / np.sqrt(p *(1-p)/n),5)
print(z)

#p-value 계산
from scipy.stats import norm
p_value = round(1- norm.cdf(z),3)
print(p_value)
```
