#### 작업형 2유형
백화점 고객의 1년 간 구매 데이터를 활용해
* 데이터 전처리
* Feature Engineering
* 모델링 (분류 알고리즘 사용)
* 하이퍼파라미터 튜닝 (초매개변수 최적화)
* 모형 앙상블
* csv제출

```python
import pandas as pd

X = pd.read_csv('/content/X_train.csv', encoding="euc-kr")
y = pd.read_csv('/content/y_train.csv', encoding="euc-kr")
test = pd.read_csv('/content/X_test.csv', encoding="euc-kr")
```

```python
# EDA
X.shape , y.shape, test.shape
```

```python
X.head()
```

```python
X.isnull().sum()
```

```python
X.info()
```

```python
# 결측치 처리
X = X.fillna(0)
test = test.fillna(0)

X = X.drop(['cust_id'], axis=1)
cust_id = test.pop('cust_id')
```

```python
# 범주형 변수 label encoding
from sklearn.preprocessing import LabelEncoder

cols = ['주구매상품', '주구매지점']
for col in cols:
  le = LabelEncoder()
  X[col] = le.fit_transform(X[col])
  test[col] = le.transform(test[col])

X.head()
```

```python
# 모델링
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X,y['gender'])
print(model.score(X, y['gender']))
predictions = model.predict_proba(test)
```

```python
output = pd.DataFrame({'cust_id': cust_id, 'gender': predictions[:,1]})

output.head()
```

```python
output.to_csv("123456789.csv", index=False)
pd.read_csv("123456789.csv")
```
